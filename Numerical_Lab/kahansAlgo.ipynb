```python
import numpy as np
```

```python
n=30000
exponents=np.random.uniform(-10,10,n)
signs=np.random.choice([-1,1],n)
data=signs*(10.0**exponents)
```

```python
print(data.min())
print(data.max())
len(data)
```

```python
sum_original = np.sum(data)

sorted_increasing = np.sort(data)
sum_increasing = np.sum(sorted_increasing)

sorted_decreasing = np.sort(data)[::-1]
sum_decreasing = np.sum(sorted_decreasing)
```

```python
print(sum_original)
print(sum_increasing)
print(sum_decreasing)
```

```python
diff_incr=abs(sum_increasing-sum_original)
diff_decr=abs(sum_decreasing-sum_original)

print(diff_incr)
print(diff_decr)
```

```python
#Percentile Rank Calculation:
def check_percentile_interactive(data):
    n = len(data)
    user_val_str = input("Enter a number: ")
    user_val = float(user_val_str)
    count_below = np.sum(data < user_val)
    percentile = (count_below / n) * 100
    print(f"Your percentile rank is: {percentile}")
```

```python
#Kahans Algo:
def kahan_sum(arr):
    sum_value=0.0
    c=0.0
    for num in arr:
        y=num-c
        t=sum_value+y
        c=(t-sum_value)-y
        sum_value=t
    return sum_value
```

```python
def manual_parcentile_rank(arr,val):
    count_below=0
    n=len(arr)
    for x in arr:
        if x<val:
            count_below+=1
    return (count_below/n)*100
```

```python
print(f"Original Sum:{sum_original}")
print(f"sum_increasing:{sum_increasing}")
print(f"sum_decreasing:{sum_decreasing}")
print(f"kahan_Sum:{kahan_sum(data)}")
print(f"kahan_sum_increased:{kahan_sum(sorted_increasing)}")
print(f"kahan_sum_decreased:{kahan_sum(sorted_decreasing)}")
print(check_percentile_interactive(data))
print(f"Manual Percentile:{manual_parcentile_rank(data,5)}")
```
