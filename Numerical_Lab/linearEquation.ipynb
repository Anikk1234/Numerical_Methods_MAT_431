```python
import numpy as np
```

```python
A=np.array([
    [3,-4,1,5],
    [1,3,-2,5],
    [-2,-3,2,2],
    [-1,5,3,4],
])
```

```python
B=np.array([7,12,2,17])
```

```python
try:
    solution = np.linalg.solve(A, B)

    # Unpack the solution
    a, b, c, d = solution

    # Format the output as requested
    print(f"a = {a:.2f} , b = {b:.2f} , c = {c:.2f} , d = {d:.2f}")

except np.linalg.LinAlgError:
    print("The system implies a singular matrix and cannot be solved uniquely.")
```

```python
rand_A = np.random.randint(-10, 10, size=(4, 4))
rand_B = np.random.randint(-10, 10, size=4)
```

```python
print("Matrix A:", rand_A)
print("Vector B:", rand_B)
```

```python
try:
    rand_solution = np.linalg.solve(rand_A, rand_B)
    ra, rb, rc, rd = rand_solution

    print("\nSolution for Random System:")
    print(f"a = {ra:.2f} , b = {rb:.2f} , c = {rc:.2f} , d = {rd:.2f}")

except np.linalg.LinAlgError:
    print("\nThe generated random matrix was singular. Please run again.")
```
